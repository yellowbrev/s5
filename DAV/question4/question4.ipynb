```python
import pandas as pd
df1=pd.read_excel('Question4_excel1.xlsx')
df2=pd.read_excel('Book2.xlsx')
```

```python
print(df1)
print("=========================================================")
print(df2)
```

Question 1

Perform merging of the two dataframes to find the names of students who had attended the
workshop on both days.


```python
df3=pd.merge(df1,df2, on='Name')
df3
```

Question 2

Find names of all students who have attended workshop on either of the days.

```python
df4=pd.merge(df1,df2,on="Name",how="outer")
df4
```

Question 3

Merge two data frames row-wise and find the total number of records in the data frame

```python
df8=pd.merge(df1,df2,how='outer')
print(df8)
print("===========================================================")
print('Total No of records is :',df8.shape[0])
```

Question 4

Merge two data frames and use two columns names and duration as multi-row indexes. Generate
descriptive statistics for this multi-index.

```python
df5=df1.set_index(["Name", "Duration"])
df6=df2.set_index(["Name", "Duration"])
df7=pd.merge(df5, df6, left_index=True, right_index=True, how='outer')
print(df7)
df7.describe()
```
