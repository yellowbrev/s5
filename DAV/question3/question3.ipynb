```python
import pandas as pd
import numpy as np
```

Create a dataframe with 3 columns and 50 rows of random numbers and replace 10% of the values with NaN

```python
data = {'A': np.random.rand(50),
        'B': np.random.rand(50),
        'C': np.random.rand(50)}
df = pd.DataFrame(data)

null_indices = np.random.choice(df.index, size=15, replace=False)
null_columns = np.random.choice(df.columns, size=15)

# Replace the values at the null_indices,null_columns by NaN;
# null_indices and null_column contain random 15 index and column resp.
for i,j in zip(null_indices,null_columns):
    df.loc[i,j] = np.nan

print(df)
```

a. Identify and count missing values in a dataframe

```python

print(df.isnull())
print(df.isnull().sum())
```

b. Drop the column having more than 5 null values

```python

#thresh specified miniumum number of non null values each column must have.
df = df.dropna(thresh=len(df)-5, axis=1)
df
```

c. Identify the row label having maximum of the sum of all values in a row and drop that row

```python
# df.sum(axis=1) will provide sum of each row of df
# idxmax will provide index of maximum value among them
df = df.drop(df.sum(axis=1).idxmax())
df
```

d. Sort the dataframe based on the first column

```python
df = df.sort_values(by='A')
df
```

e. Remove all duplicates from the first column

```python
df = df.drop_duplicates(subset='A')
df
```

f. Find correlation between the first and second column, covariance between second and third column

```python
correlation = df['A'].corr(df['B'])
covariance = df['B'].cov(df['C'])
print("Correlation between Column1 and Column2:", correlation)
print("Covariance between Column2 and Column3:", covariance)
```

g. Detect outliers and remove rows with outliers

```python
# df.mean() and df.std() give mean and std of each column
z_scores = (df - df.mean()) / df.std()


# z score is a way to check outliers.
# non outliers con
threshold=1.25
outliers = (z_scores < -threshold) | (z_scores > threshold)


# outliers.any(axis=1) will return True if any one element in the row is True i.e outliers
# ~outliers.any(axis=1) will return True If all elements contain False in the row i.e. not even a single outliers 
df=df[~outliers.any(axis=1)]
df
```

h. Discretize the second column and create 5 bins

```python
df['B_bins'] = pd.cut(np.array(df['B']), bins=5)
df
```
